```
def avg(data):
    x = sum(data) / len(data)
    return x

def cleared_data(data):
    for i in data:
       if i is None: 
          data.remove(i)
    return data      

n = int(input('Количество измерений: '))

Data = list()
for i in range(n):
    d = input('Измерениt: '.format(i+1))
    Data.append(None if d == '-' 
                   else float(d))  

Data1 = cleared_data(Data)
print('Средняя: {0:.2f}'.format(avg(Data1)))

# --------------
# Пример вывода:
#
# Кол-во измерений: 3
# Измерение 1-е: 10
# Измерение 2-е: -
# Измерение 3-е: 20
# Средняя температура: 15.00
```

```
def is_lucky(num):
    num_even_count = 0
    num_odd_count = 0

    while num != 0:
        if num % 2 == 0:
           num_even_count += 1
        else:
           num_odd_count += 1  
        num = num // 10

    return True if num_even_count == num_odd_count else False  

def lucky_numbers(a, b):
    num = list()
    for i in range(a, b + 1):
        if is_lucky(i):
           num.append(i) 
    return num  

a = int(input('Первый номер билета: '))
b = int(input('Последний номер билета: '))

num = lucky_numbers(a, b)
for i in num:
    print(i, end = ' ')

# --------------
# Пример вывода:
#
# Первый номер билета: 10
# Последний номер билета: 25
# 10 12 14 16 18 21 23 25
```

```
def is_leap(year):
    if year % 4 == 0:
        return True
    else:
        return False


def days(month, year):
    if month in (1,3,5,7,8,10,12):
        return 31
    elif month == 2:
        if is_leap(year):
            return 29
        else:
            return 28    
    else: 
        return 30


def another_date(day, month, year, delta=1):

    def previous_date(day, month, year):
        if day == 1:
            if month == 1:
                month = 12
                year -= 1
            else:     
                month -= 1 
            day =  days(month, year)    
        else:
            day -= 1    

        return day, month, year

    def next_date(day, month, year):
        if day == days(month, year):
            day = 1
            if month == 12:
                month = 1
                year += 1
            else:     
                month += 1 
        else:
            day += 1   

        return day, month, year  

    if delta > 0:
        for i in range(delta):
           day, month, year = next_date(day, month, year)
    else:
        for i in range(0, delta, -1):
           day, month, year = previous_date(day, month, year)   
    return day, month, year       

    
date = input('День, месяц, год через пробел: ').split(sep=' ')
day = int(date[0]) 
month = int(date[1])
year = int(date[2])

delta = int(input('Свдиг: '))
day, month, year = another_date(day, month, year, delta)
print('Новый день: {:0>2}.{:0>2}.{:0>4}'.format(day, month, year))

# --------------
# Пример вывода:
#
# День, месяц, год через пробел: 1 1 2000
# Свдиг (может быть отрицательным): -2
# Новый день: 30/12/1999
#
# День, месяц, год через пробел: 1 1 2000
# Свдиг (может быть отрицательным): 2
# Новый день: 03/01/2000
```

```
def gcd(first, second):
    while first != 0 and second != 0:
        if first > second:
            first = first % second
        else:  
            second = second % first
    return first + second          


def lcm(first, second):
    if gcd(first, second) != 0:
        return (abs(first * second))/gcd(first, second)
    else:
        return 0
  

def gcd_nums(nums):
    result = nums[0]
    for i in range(1, len(nums)):
        result = gcd(result, nums[i])
    return result
    
def lcm_nums(nums):
    result = nums[0]
    for i in range(1, len(nums)):
        result = lcm(result, nums[i]) 
    return result

nums = [int(i) for i in input('Введите числа через пробел: ').split(sep=' ')]


nod = gcd_nums(nums)
nok = lcm_nums(nums)

print('НОД = ', nod)
print('НОК = ', nok)
# --------------
# Пример вывода:
#
# Введите числа через пробел: 8 10 14
# НОД = 2
# НОК = 280
#
# Введите числа через пробел: 6 8 24 16
# НОД = 2
# НОК = 48
```

```
def has_digits(sentence):
    for i in sentence: 
        if i in ['1', '2', '3', '4', '5', '6', '7', '8', '9']:
            return True
        
    return False

def sentences_with_digits_count(sentences):
    count = 0
    for i in sentences:
        if has_digits(i):
            count += 1 
    return count

n = int(input('Введите количество предложений: '))

sentences = list()
for i in range(0, n):
    sentence = input('Введите предложение № {}: '.format(i+1))
    sentences.append(sentence)
t = sentences_with_digits_count(sentences)
print('Предложений с цифрой = ', t)
# --------------
# Пример вывода:
#
# Введите количество предложений: 3
# Введите предложение №1:
# Просто текст
# Введите предложение №2:
# Текст с цифрой 1 (один)
# Введите предложение №3:
# Тут нет цифры
# Предложений с цифрой = 1
```

```
def sentence_stats(sentence):
    d = dict()
    sentence = sentence.lower()
    for i in sentence:
        d[i] = d.get(i, 0) + 1
    return d

s = input('Введите предложение: ')
print(sentence_stats(s))
# --------------
# Пример вывода:
#
# Введите предложение: мама МЫла РамУ
# {'л': 1, 'р': 1, 'у': 1, 'м': 4, 'а': 4, 'ы': 1, ' ': 2}
```
